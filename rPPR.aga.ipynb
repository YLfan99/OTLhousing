```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

import xgboost as xgb
import lightgbm as lgb

np.random.seed(42)

from utils import *
```

```python
data = pd.read_csv('data/ver2_house_sold_history.csv')
data
```

```python
data_train = data[(data['Year']<2017)].drop(columns=['Year','Month','PropID','Suburb'])
data_test = data[(data['Year']>2016)].drop(columns=['Year','Month','PropID','Suburb'])
X_train, X_dev, y_train, y_dev = train_test_split(data_train.drop(columns=['Price']),data_train['Price'],test_size=0.2)
X_test, y_test = data_test.drop(columns=['Price']),data_test['Price']
model = lgb.LGBMRegressor(num_leaves=500, n_estimators=1500, reg_alpha=0, max_depth=-1)
model.fit(X_train, y_train, eval_set=[(X_dev, y_dev)])
```

```python
params_lgb = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'learning_rate': 0.1,
    'num_leaves': 50,
    'max_depth': 6,
    'subsample': 0.8,
}
```

```python
train_data_lgb = lgb.Dataset(data.drop(columns=['Year','Month','PropID','Suburb']),data['Price'])
cv_results = lgb.cv(
    params_lgb, train_data_lgb, num_boost_round=10000, nfold=5, early_stopping_rounds=500, verbose_eval=50, show_stdv=True, stratified=False, shuffle=True, metrics='l2', seed=42
)
print('best n_estimators', len(cv_results['l2']))
```

```python
model_lgb = lgb.LGBMRegressor(objective='regression', num_leaves=50, learning_rate=0.02, n_estimators=1988, metric='mse', bagging_fraction=0.8, feature_fraction=0.8)

params_test = {
    'num_leaves': range(50,1002,50),
}

gs = GridSearchCV(estimator=model_lgb, param_grid=params_test, scoring='neg_mean_squared_error', cv=5, verbose=1, n_jobs=4)
gs.fit(data.drop(columns=['Year','Month','PropID','Suburb']),data['Price'])
print('best parameters:', gs.best_estimator_)
```
